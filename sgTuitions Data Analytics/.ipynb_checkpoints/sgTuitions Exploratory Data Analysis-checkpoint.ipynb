## Objective 1: Composition of Tutors

We begin by choosing the columns which we are interested in for this analysis, they are: ["Telegram ID","Level of Study","Rates","Gender","Occupation"].

After which, I replaced all single quotes and hyphens with empty spaces as there are many variants of "A Levels" under the "Level of Study" column, such as, "alvls", "a-lvls", "a'lvls", "'a'levels" and many others.

Then, from the .contains() method, one can get a boolean mask (store it in "Cleaned Level of Study") and via the .mask() method, I replaced the string in Level of Study column with that in the corresponding Cleaned Level of Study column, e.g. "a-lvls" is being replaced by "A Levels".

This is repeated to search for other Categories, namely: Graduates, Undergraduates, A Levels, Diploma, O Levels

### Lessons Learnt
1. The distinction between Undergraduates and A Levels/Diploma is ambiguous (because most undergraduates have either A Level certificate or a diploma and if Categories are set up via the latest attained certificate, this distinction is then ill-defined).
2. Also, in the first place, data collection can be better managed so that the data cleaning process would not be tedious. Instead of letting applicants freely type their education level, the bot could provide a list of education levels. 

```python
# Some stuff to make the work easier.
tutorsdfsimplified = tutorsdf[["Telegram ID", "Level of Study", "Rates",
                               "Gender", "Occupation", "Experiences", "Remarks", "NumRates"]]
tutorsdfsimplified["Cleaned Level of Study"] = 'None'
tutorsdfsimplified["Level of Study"] = tutorsdfsimplified["Level of Study"].replace(
    to_replace={r"\â€™": r" ", r"\-": r" "}, regex=True)
tutorsdfsimplified["Level of Study"] = tutorsdfsimplified["Level of Study"].str.lower()

# Find Graduates in Level of Study by searching some keywords like Degree, Bachelor.
Grads = tutorsdfsimplified["Level of Study"].str.contains(
    r".*degre.*|.*b.ch.l.*|master.*|.*ph\.d.*|.*phd.*|\sgrad.*$|^grad.*$", na=False, regex=True)
tutorsdfsimplified["Cleaned Level of Study"] = tutorsdfsimplified.loc[:,
                                                                      "Cleaned Level of Study"].mask(Grads,
                                                                                                     other='Graduates')

# Find Undergraduates in Level of Study by searching 'Undergrad...' and 'University'.
Undergrads = tutorsdfsimplified["Level of Study"].str.contains(
    r'.*undergrad.*|.*uni.*', na=False, regex=True)

# Use the .mask() method to replace whatever that is True in the Undergrad mask to the corresponding element
# in the 'Clean Level of Study' column.
tutorsdfsimplified["Cleaned Level of Study"] = tutorsdfsimplified.loc[:,
                                                                      "Cleaned Level of Study"].mask(Undergrads,
                                                                                                     other='Undergraduates')

# Find A Levels in Level of Study by searching 'A Level' and the likes.
Alevel = tutorsdfsimplified["Level of Study"].str.contains(
    r".*a\sl.v.l.*|.*a\slvl|.*a-lvl|.*a'lvl|.*alvl|.*alevel.*", na=False, regex=True)
tutorsdfsimplified["Cleaned Level of Study"] = tutorsdfsimplified.loc[:,
                                                                      "Cleaned Level of Study"].mask(Alevel,
                                                                                                     other='A Levels')

# Find O Levels in Level of Study by searching 'O Level' and the likes.
Olevel = tutorsdfsimplified["Level of Study"].str.contains(
    r".*o\sl.v.l.*|.*o\slvl|.*o-lvl|.*o'lvl|.*olvl|.*olevel.*", na=False, regex=True)
tutorsdfsimplified["Cleaned Level of Study"] = tutorsdfsimplified.loc[:,
                                                                      "Cleaned Level of Study"].mask(Olevel,
                                                                                                     other='O Levels')

# Find Polygraduates in Level of Study.
Poly = tutorsdfsimplified["Level of Study"].str.contains(
    r".*diploma.*|.*poly.*", na=False, regex=True)
tutorsdfsimplified["Cleaned Level of Study"] = tutorsdfsimplified.loc[:,
                                                                      "Cleaned Level of Study"].mask(Poly,
                                                                                                     other='Diploma')
```

When I look at the tutorsdfsimplified DataFrame, I realized some tutors put their Level of Study as A Levels while their Occupation is listed as Undergraduates. In light of this, I decided to use the Occupation column to augment the assignment of their Level of Study.

```python
# Some codes to simplify the work.
tutorsdfsimplified["Occupation"] = tutorsdfsimplified["Occupation"].replace(to_replace={r"\â€™": r" ", r"\-": r" "}, regex=True)
tutorsdfsimplified["Occupation"] = tutorsdfsimplified["Occupation"].str.lower()

# Find Undergraduates in Occupation
UndergradsbyOccu = tutorsdfsimplified["Occupation"].str.contains(r'.*undergrad.*|.*uni.*|^undergrad.*', na=False, regex=True)
tutorsdfsimplified["Cleaned Level of Study"] = tutorsdfsimplified.loc[:,"Cleaned Level of Study"].mask(UndergradsbyOccu,
                                                                                                       other='Undergraduates')

# Check how many are still unassigned.
print("Length of None is {}".format(len(tutorsdfsimplified[tutorsdfsimplified["Cleaned Level of Study"] == 'None'])))


# Finally, replace the Level of Study column by the Cleaned Level of Study
tutorsdfsimplified["Level of Study"] = tutorsdfsimplified["Cleaned Level of Study"]
tutorsdfsimplified = tutorsdfsimplified.drop(["Cleaned Level of Study"], axis=1)
```

Here we see the DataFrame with the columns which we are interested in, namely: Tutors' Level of Study, Their Desired Rates and Gender.

```python
tutorsdfsimplified[["Level of Study", "Rates", "Gender"]].head()
```

Let us find out how many of the tutors are MOE-trained.

```python
tutorsdfsimplified[["Telegram ID", "Level of Study","Rates", "Gender"]].groupby("Gender").size()
tutorsdfsimplified[["Telegram ID", "Level of Study","Rates", "Gender"]].groupby("Rates").size()
Dist_LoS = tutorsdfsimplified[["Telegram ID", "Level of Study", "Rates", "Gender"]].groupby(["Level of Study"]).size()
(100. * Dist_LoS / Dist_LoS.sum()).round(1).astype(str) + '%'
```

```python
tutorsdfsimplified["MOE-Trained"] = 'Not MOE-Trained'

# Find out if the tutors are MOE trained by Occupation
MOEbyOccu = tutorsdfsimplified["Occupation"].str.contains(r'.*moe.*|.*moe.*|^moe.*|moe', na=False, regex=True)
tutorsdfsimplified["MOE-Trained"] = tutorsdfsimplified.loc[:,"MOE-Trained"].mask(MOEbyOccu, other='MOE-Trained')

tutorsdfsimplified["Experiences"] = tutorsdfsimplified["Experiences"].replace(to_replace={r"\â€™": r" ", r"\-": r" "},
                                                                              regex=True)
tutorsdfsimplified["Experiences"] = tutorsdfsimplified["Experiences"].str.lower()

# Find out if the tutors are MOE trained by Experiences
MOEbyExp = tutorsdfsimplified["Experiences"].str.contains(r'.*moe.*|.*moe.*|^moe.*', na=False, regex=True)
tutorsdfsimplified["MOE-Trained"] = tutorsdfsimplified.loc[:,"MOE-Trained"].mask(MOEbyExp, other='MOE-Trained')

tutorsdfsimplified.loc[tutorsdfsimplified['Level of Study'] == 'Undergraduates', "MOE-Trained"] = 'Not MOE-Trained'

tutorsdfsimplified.groupby('MOE-Trained').size()*100/len(tutorsdfsimplified)
```

From the above result, we can estimate around 5.2% of the tutor has some form of MOE-teaching training.

```python
tutorsdfsimplified["Full-Time"] = 'Not Full-Time'

# Find out if the tutors are Full-Time tutors by Occupation
FTbyOccu = tutorsdfsimplified["Occupation"].str.contains(r'.*full time.*|.*full time.*|^full time.*', na=False, regex=True)
tutorsdfsimplified["Full-Time"] = tutorsdfsimplified.loc[:,"Full-Time"].mask(FTbyOccu, other='Full-Time')

tutorsdfsimplified["Experiences"] = tutorsdfsimplified["Experiences"].replace(to_replace={r"\â€™": r" ", r"\-": r" "},
                                                                              regex=True)
tutorsdfsimplified["Experiences"] = tutorsdfsimplified["Experiences"].str.lower()

# Find out if the tutors are Full-Time tutors by Experiences
FTbyExp = tutorsdfsimplified["Experiences"].str.contains(r'.*full time tutor.*|.*full time tutor.*|^full time tutor.*',
                                                         na=False, regex=True)
tutorsdfsimplified["Full-Time"] = tutorsdfsimplified.loc[:,"Full-Time"].mask(FTbyExp, other='Full-Time')

tutorsdfsimplified.loc[tutorsdfsimplified['Level of Study'] == 'Undergraduates', "Full-Time"] = 'Not Full-Time'

tutorsdfsimplified.groupby('Full-Time').size()*100/len(tutorsdfsimplified)
```

From the above result, we can estimate around 15.2% of the tutors are full-time tutors.

## Objective 2: Understand the Distribution of Rates Quoted according to Tutors' Attributes

It would be interesting to do a multilevel GroupBy to see how the quoted Rates are distributed across Level of Study, whether the tutor is self-declared to be MOE-trained and whether the tutor is a self-declared full-time tutor.

```python
LOS_Rates = tutorsdfsimplified[[
    "Telegram ID", "Level of Study", "Rates", "Gender"]].groupby(["Level of Study", "Rates"]).size()
print(LOS_Rates[0:3])
MOET_Rates = tutorsdfsimplified[["Telegram ID", "MOE-Trained", "Rates", "Gender"]].groupby(["MOE-Trained", "Rates"]).size()
print(MOET_Rates[0:3])
FT_Rates = tutorsdfsimplified[["Telegram ID", "Full-Time", "Rates", "Gender"]].groupby(["Full-Time", "Rates"]).size()
print(FT_Rates[0:3])
```

Here, we obtain a MultiIndex DataFrame which are sorted according to first the Level of Study, Gender and their desired Rates.

Then we use a Seaborn Categorical plot to look at the bar plot of the Rates quoted distribution across the different educational levels.

```python
LOS_Rates = tutorsdfsimplified[["Telegram ID", "Level of Study","Rates",
                                "MOE-Trained", "Full-Time"]].groupby(["Level of Study", "Rates"],
                                                                     as_index=False).size().reset_index()
print(LOS_Rates.columns)
LOS_Rates.rename(columns={0: 'Count'}, inplace=True)
LOS_Rates.drop(LOS_Rates[LOS_Rates['Rates'] == 'UseSavedProfile'].index,inplace=True)
sns.set(style="whitegrid", font_scale=5)
sns.catplot(x='Rates', y='Count',
            row='Level of Study', data=LOS_Rates,
            kind='bar', height=10, aspect=5)
```

```python
MOET_Rates = tutorsdfsimplified[["Telegram ID", "Level of Study",
                                 "Rates", "MOE-Trained", "Full-Time"]].groupby(["MOE-Trained",
                                                                                "Rates"],
                                                                               as_index=False).size().reset_index()
MOET_Rates.rename(columns={0: 'Count'}, inplace=True)
MOET_Rates.drop(MOET_Rates[MOET_Rates['Rates'] == 'UseSavedProfile'].index, inplace=True)
sns.set(style="whitegrid", font_scale=5)
sns.catplot(x='Rates', y='Count',
            row='MOE-Trained', data=MOET_Rates,
            kind='bar', height=10, aspect=5)
```

```python
FT_Rates = tutorsdfsimplified[["Telegram ID", "Level of Study","Rates",
                               "MOE-Trained", "Full-Time"]].groupby(["Full-Time", "Rates"],
                                                                    as_index=False).size().reset_index()
FT_Rates.rename(columns={0: 'Count'}, inplace=True)
FT_Rates.drop(FT_Rates[FT_Rates['Rates'] == 'UseSavedProfile'].index, inplace=True)
sns.set(style="whitegrid", font_scale=5)
sns.catplot(x='Rates', y='Count',
            row='Full-Time', data=FT_Rates,
            kind='bar', height=10, aspect=5)
```

## Ideas for Further Analysis

### Background

1. The Column "Notified" has 3 labels, a. Rejected - meaning the tutor is rejected by the job poster (tuition agent, parent or student). b. Shortlisted - meaning the tutor is preliminary accepted by the job poster and their Telegram contacts are exchanged. c. Notified - meaning the tutor is neither shortlisted nor rejected, it is the default state.


2. The number of Shortlist a job poster can shortlist tutors for is 5 per assignment. A job poster has a weekly limit of assignment that he can post on the channel. Essentially, the number of tutors a job poster can shortlist is finite. Given the aim of using sgTuitions is to find a suitable tutor, and no personal information is conveyed to the job poster before he shortlists a tutor except all that is provided to the bot, we have strong reason to believe job poster does read whatever that is on a profile and shortlists a tutor with the intention of hiring, short of coordinating the nitty-gritty details like tuition timing and so on.

### Future Idea

To predict, given the features of the application, e.g. Rates quoted, educational level, full-time tutor or otherwise, if a particular tutor will be shortlisted.

The more important features could be: Rates quoted, Experiences and Remarks. However, as Experiences and Remarks are long strings, it may not be easy to analyze them.
